```python
import os
import pandas as pd
import hddm
```

```python
# Load data
data = hddm.load_csv('ddm_combine.csv')
data.rt = data.rt/1000
data.response = 1-data.response
data = hddm.utils.flip_errors(data)
```

```python
data.head(10)
```

```python
data = data[data.stimulus!='tree']
```

```python
data
```

```python
m = hddm.HDDM(data, depends_on={'v': ['type','subj_race'], 'a': ['type','subj_race'], 'z': ['type','subj_race']}, bias = True, p_outlier=.05)
m.find_starting_values()
m.sample(20000, burn=1000, dbname='db_combine_notree', db='pickle')
m.save('model_combine_notree')
```

```python
m = hddm.HDDM(data, depends_on={'v': ['type','c_race'], 'a': ['type','c_race'], 'z': ['type','c_race']}, bias = True, p_outlier=.05)
m.find_starting_values()
m.sample(20000, burn=1000, dbname='db_combine_new', db='pickle')
m.save('model_combine_new')
```
