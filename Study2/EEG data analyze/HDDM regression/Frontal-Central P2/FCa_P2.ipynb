```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import hddm
```

```python
# Load data
data = hddm.load_csv('eeg_ddm_StimAllElec_new.csv')
data.rt = data.rt/1000
data.response = 1-data.response
data = hddm.utils.flip_errors(data)
data = data[data['type']=='pain']
data = data[data['race']=='Asian']
data
```

```python
data['FCa_P2'] = data[['FC1_P2','FC2_P2','FC3_P2','FC4_P2','FCz_P2']].mean(axis=1)
data
```

```python
m_reg_new = hddm.HDDMRegressor(data,
                           "a ~ FCa_P2",
                           depends_on={'v': 'race'})
m_reg_new.sample(20000, burn=1000, dbname='db_FCa_P2', db='pickle')
m_reg_new.save('model_FCa_P2')
```

```python
a_FCa_P2 = m_reg_new.nodes_db.node["a_FCa_P2"]
```

```python
hddm.analyze.plot_posterior_nodes([a_FCa_P2], bins=20)
plt.xlabel('a_FCa_P2 coeffecient in ')
print("P( > 0) = ", (a_FCa_P2.trace() < 0).mean())
```
