```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import hddm
```

```python
# Load data
data = hddm.load_csv('eeg_ddm_StimAllElec_new.csv')
data.rt = data.rt/1000
data.response = 1-data.response
data = hddm.utils.flip_errors(data)
data = data[data['type']=='pain']
data = data[data['race']=='Asian']
data
```

```python
data['Ca_N2'] = data[['C1_N2','C2_N2','C3_N2','C4_N2','Cz_N2']].mean(axis=1)
data
```

```python
m_reg_new = hddm.HDDMRegressor(data,
                           "a ~ Ca_N2",
                           depends_on={'v': 'race'})
m_reg_new.sample(20000, burn=1000, dbname='db_Ca_N2', db='pickle')
m_reg_new.save('model_Ca_N2')
```

```python
a_Ca_N2 = m_reg_new.nodes_db.node["a_Ca_N2"]
```

```python
hddm.analyze.plot_posterior_nodes([a_Ca_N2], bins=20)
plt.xlabel('a_Ca_N2 coeffecient in ')
print("P( > 0) = ", (a_Ca_N2.trace() < 0).mean())
```
